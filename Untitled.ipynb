```python
ann = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=10, activation='relu'),
    tf.keras.layers.Dense(units=9, activation='relu'),
    tf.keras.layers.Dense(units=4, activation='relu'),
    tf.keras.layers.Dense(units=4, activation='relu'),
    tf.keras.layers.Dense(units=2, activation='relu'),

    tf.keras.layers.Dense(units=1, activation='sigmoid')
])

# Compile the model
ann.compile(optimizer='adam', loss=tf.keras.losses.BinaryCrossentropy(), metrics=[tf.keras.metrics.BinaryAccuracy()])

# Train the model with validation data
history = ann.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size=50, epochs=100)

# Make predictions
y_pred_ann = ann.predict(x_test)
y_pred_binary = (y_pred_ann > 0.5).astype(int)

# Compute accuracy
accuracy = accuracy_score(y_test, y_pred_binary)
print("Test Accuracy:", accuracy)

# Evaluate training and validation accuracy
train_accuracy = history.history['binary_accuracy']
val_accuracy = history.history['val_binary_accuracy']

# Plot training and validation accuracy
import matplotlib.pyplot as plt

epochs = range(1, len(train_accuracy) + 1)

plt.plot(epochs, train_accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```
